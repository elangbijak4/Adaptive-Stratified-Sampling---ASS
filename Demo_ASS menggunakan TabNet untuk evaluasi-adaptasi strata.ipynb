```python
# ===== Cell 1: Install dependencies (Colab) =====
# Jalankan ini sekali di Colab. Jika Anda sudah menginstal pytorch-tabnet, boleh di-skip.
# Perhatian: pada runtime Colab, pilih "Runtime > Change runtime type" -> GPU jika tersedia.
!pip install -q pytorch-tabnet==0.3.1 torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
# catatan: versi dan index-url bisa disesuaikan jika GPU / CUDA berbeda. Jika gagal, coba tanpa index-url:
# !pip install -q pytorch-tabnet

# ===== Cell 2: Imports & dataset synthetic =====
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from pytorch_tabnet.tab_model import TabNetClassifier
import torch
import random

np.random.seed(42)
random.seed(42)
torch.manual_seed(42)

# Build synthetic dataset (sama seperti demo sebelumnya)
n = 8000
X = np.random.normal(size=(n, 6))
strata_probs = [0.6, 0.2, 0.15, 0.05]
strata = np.random.choice(len(strata_probs), size=n, p=strata_probs)

weights = np.array([0.5, -0.3, 0.2, 0.1, -0.2, 0.3])
base_score = X.dot(weights)
strata_intercepts = np.array([-0.2, 0.1, 0.5, -0.5])
score = base_score + strata_intercepts[strata]
probs = 1 / (1 + np.exp(-score))
y_clean = (probs > 0.5).astype(int)

label_noise_by_stratum = {0: 0.02, 1: 0.05, 2: 0.12, 3: 0.25}
y = y_clean.copy()
for s, noise in label_noise_by_stratum.items():
    idx = np.where(strata == s)[0]
    flip = np.random.rand(len(idx)) < noise
    y[idx[flip]] = 1 - y[idx[flip]]

df = pd.DataFrame(X, columns=[f"f{i}" for i in range(X.shape[1])])
df["stratum"] = strata
df["y"] = y

train_df, val_df = train_test_split(df, stratify=df["stratum"], test_size=0.25, random_state=42)

print("Train composition by stratum:\n", train_df["stratum"].value_counts().sort_index())

# ===== Cell 3: ASS helper functions =====
unique_strata = sorted(df["stratum"].unique())
n_strata = len(unique_strata)
strata_counts = train_df["stratum"].value_counts().sort_index()

# initialize uniform weights
weights = pd.Series(1.0 / n_strata, index=unique_strata).astype(float)

def sample_train_by_weights(train_df, strata_weights, sample_size=None, random_state=None):
    if sample_size is None:
        sample_size = len(train_df)
    rng = np.random.RandomState(random_state)
    base_counts = train_df["stratum"].value_counts().sort_index()
    desired_score = strata_weights * base_counts
    desired_share = desired_score / desired_score.sum()
    desired_counts = np.round(desired_share * sample_size).astype(int)
    for s in unique_strata:
        if desired_counts[s] < 1:
            desired_counts[s] = 1
    pieces = []
    for s in unique_strata:
        subset = train_df[train_df["stratum"] == s]
        k = desired_counts[s]
        chosen = subset.sample(n=k, replace=True, random_state=rng.randint(0, 1_000_000))
        pieces.append(chosen)
    sampled = pd.concat(pieces).sample(frac=1.0, random_state=rng.randint(0, 1_000_000)).reset_index(drop=True)
    return sampled

def per_stratum_error(model, X_val_df, y_val_series):
    # model.predict_proba expects numpy arrays
    preds = model.predict(X_val_df.values)
    dfp = X_val_df.copy()
    dfp["pred"] = preds
    dfp["y"] = y_val_series.values
    # error = fraction incorrect
    err = dfp.groupby("stratum").apply(lambda g: (g["pred"] != g["y"]).mean()).to_dict()
    return err

# ===== Cell 4: ASS + TabNet training loop =====
# Hyperparameters for ASS update rule
alpha = 0.85  # weight for error-driven update
beta = 0.15   # weight for inverse-frequency baseline
momentum = 0.6

EPOCHS = 10
history = {"epoch": [], "acc": [], "f1": [], "weights": [], "per_stratum_error": []}

# Prepare validation X,y
X_val_df = val_df[[c for c in val_df.columns if c.startswith("f")]].reset_index(drop=True)
X_val_df["stratum"] = val_df["stratum"].values  # keep stratum column for per-stratum grouping
y_val = val_df["y"].reset_index(drop=True)

# We'll configure a simple TabNet model. TabNetClassifier accepts numpy X,y.
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("Using device:", device)

for epoch in range(1, EPOCHS+1):
    # 1) sample training set according to current weights
    sampled = sample_train_by_weights(train_df, weights, sample_size=len(train_df), random_state=epoch*31)
    X_train = sampled[[c for c in sampled.columns if c.startswith("f")]].values
    y_train = sampled["y"].values.astype(int)

    # 2) instantiate and train TabNetClassifier (retrain each epoch for demo simplicity)
    clf = TabNetClassifier(
        n_d=8, n_a=8, n_steps=5,
        optimizer_params=dict(lr=2e-2),
        mask_type='sparsemax',
        verbose=0,
        device_name=device
    )
    # TabNet wants numpy arrays; we can pass max_epochs small for speed
    clf.fit(
        X_train, y_train,
        eval_set=[(X_train, y_train), (val_df[[c for c in val_df.columns if c.startswith("f")]].values, val_df["y"].values)],
        eval_name=['train','valid'],
        eval_metric=['accuracy'],
        max_epochs=30, patience=5,
        batch_size=256, virtual_batch_size=64
    )

    # 3) evaluate on validation set
    X_val_features = val_df[[c for c in val_df.columns if c.startswith("f")]]
    preds = clf.predict(X_val_features.values).reshape(-1)
    acc = accuracy_score(y_val, preds)
    f1 = f1_score(y_val, preds)
    prec = precision_score(y_val, preds)
    rec = recall_score(y_val, preds)

    # compute per-stratum error
    df_val_with_pred = val_df.copy().reset_index(drop=True)
    df_val_with_pred["pred"] = preds
    err = df_val_with_pred.groupby("stratum").apply(lambda g: (g["pred"] != g["y"]).mean()).to_dict()
    err_s = pd.Series([err[s] for s in unique_strata], index=unique_strata)

    # 4) update weights (blend error_norm with inverse-frequency baseline)
    inv_freq = 1.0 / strata_counts
    inv_freq_norm = inv_freq / inv_freq.sum()
    err_norm = err_s / err_s.sum()
    new_weights = alpha * err_norm + beta * inv_freq_norm
    updated = (1 - momentum) * weights + momentum * new_weights
    updated = updated / updated.sum()
    weights = updated.copy()

    # record
    history["epoch"].append(epoch)
    history["acc"].append(acc)
    history["f1"].append(f1)
    history["weights"].append(weights.copy())
    history["per_stratum_error"].append(err.copy())

    print(f"[Epoch {epoch}] acc={acc:.4f}, f1={f1:.4f}, prec={prec:.4f}, rec={rec:.4f}")
    print(" per-stratum error:", {k: round(v,3) for k,v in err.items()})
    print(" updated weights:", {int(k): round(float(weights[k]),3) for k in weights.index})
    print("-"*70)

# ===== Cell 5: Visualize results =====
hist_df = pd.DataFrame({"epoch": history["epoch"], "acc": history["acc"], "f1": history["f1"]})

err_matrix = pd.DataFrame(history["per_stratum_error"]).T
err_matrix.columns = [f"epoch_{e}" for e in history["epoch"]]
err_matrix.index = [f"stratum_{s}" for s in err_matrix.index]

weights_matrix = pd.DataFrame(history["weights"]).T
weights_matrix.columns = [f"epoch_{e}" for e in history["epoch"]]
weights_matrix.index = [f"stratum_{s}" for s in weights_matrix.index]

plt.figure(figsize=(8,4))
plt.plot(hist_df["epoch"], hist_df["acc"], marker='o')
plt.plot(hist_df["epoch"], hist_df["f1"], marker='s')
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Overall validation metrics across ASS+TabNet epochs")
plt.legend(["Accuracy","F1"])
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8,4))
for s in err_matrix.index:
    plt.plot(range(1, err_matrix.shape[1]+1), err_matrix.loc[s].values, marker='o', label=s)
plt.xlabel("Epoch")
plt.ylabel("Validation Error (per stratum)")
plt.title("Per-stratum validation error evolution")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8,4))
for s in weights_matrix.index:
    plt.plot(range(1, weights_matrix.shape[1]+1), weights_matrix.loc[s].values, marker='o', label=s)
plt.xlabel("Epoch")
plt.ylabel("Sampling Weight (normalized)")
plt.title("Adaptive strata weights across epochs")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Print final summary
final_weights = pd.DataFrame({
    "stratum": unique_strata,
    "final_weight": [weights[s] for s in unique_strata],
    "train_count": [strata_counts[s] for s in unique_strata],
    "label_noise": [label_noise_by_stratum[s] for s in unique_strata]
}).set_index("stratum")

print("\nFinal strata weights and info:")
display(final_weights)
```
