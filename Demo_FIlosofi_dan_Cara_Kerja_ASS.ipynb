```python
# Demo ASS (Colab-friendly, tanpa ace_tools)
# Salin + jalankan di Google Colab
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
import matplotlib.pyplot as plt
import random

np.random.seed(42)

# 1) Create synthetic dataset
n = 8000
X = np.random.normal(size=(n, 6))
strata_probs = [0.6, 0.2, 0.15, 0.05]
strata = np.random.choice(len(strata_probs), size=n, p=strata_probs)

weights = np.array([0.5, -0.3, 0.2, 0.1, -0.2, 0.3])
base_score = X.dot(weights)
strata_intercepts = np.array([-0.2, 0.1, 0.5, -0.5])
score = base_score + strata_intercepts[strata]
probs = 1 / (1 + np.exp(-score))
y_clean = (probs > 0.5).astype(int)

label_noise_by_stratum = {0: 0.02, 1: 0.05, 2: 0.12, 3: 0.25}
y = y_clean.copy()
for s, noise in label_noise_by_stratum.items():
    idx = np.where(strata == s)[0]
    flip = np.random.rand(len(idx)) < noise
    y[idx[flip]] = 1 - y[idx[flip]]

df = pd.DataFrame(X, columns=[f"f{i}" for i in range(X.shape[1])])
df["stratum"] = strata
df["y"] = y

train_df, val_df = train_test_split(df, stratify=df["stratum"], test_size=0.25, random_state=42)

print("Dataset composition (train):")
print(train_df["stratum"].value_counts().sort_index())

# 2) ASS algorithm (toy)
unique_strata = sorted(df["stratum"].unique())
n_strata = len(unique_strata)
strata_counts = train_df["stratum"].value_counts().sort_index()

# initialize uniform weights
weights = pd.Series(1.0 / n_strata, index=unique_strata).astype(float)

history = {"epoch": [], "overall_acc": [], "overall_f1": [], "per_stratum_error": [], "weights": []}

def sample_train_by_weights(train_df, strata_weights, sample_size=None, random_state=None):
    if sample_size is None:
        sample_size = len(train_df)
    rng = np.random.RandomState(random_state)
    base_counts = train_df["stratum"].value_counts().sort_index()
    desired_score = strata_weights * base_counts
    desired_share = desired_score / desired_score.sum()
    desired_counts = np.round(desired_share * sample_size).astype(int)
    for s in unique_strata:
        if desired_counts[s] < 1:
            desired_counts[s] = 1
    pieces = []
    for s in unique_strata:
        subset = train_df[train_df["stratum"] == s]
        k = desired_counts[s]
        chosen = subset.sample(n=k, replace=True, random_state=rng.randint(0, 1_000_000))
        pieces.append(chosen)
    sampled = pd.concat(pieces).sample(frac=1.0, random_state=rng.randint(0, 1_000_000)).reset_index(drop=True)
    return sampled

def per_stratum_error(clf, val_df):
    preds = clf.predict(val_df[[c for c in val_df.columns if c.startswith("f")]])
    dfp = val_df.copy()
    dfp["pred"] = preds
    err = dfp.groupby("stratum").apply(lambda g: (g["pred"] != g["y"]).mean()).to_dict()
    return err

EPOCHS = 12
for epoch in range(1, EPOCHS + 1):
    sampled_train = sample_train_by_weights(train_df, weights, sample_size=len(train_df), random_state=epoch*13)
    X_train = sampled_train[[c for c in sampled_train.columns if c.startswith("f")]].values
    y_train = sampled_train["y"].values
    X_val = val_df[[c for c in val_df.columns if c.startswith("f")]].values
    y_val = val_df["y"].values

    clf = RandomForestClassifier(n_estimators=80, random_state=epoch*7, n_jobs=-1)
    clf.fit(X_train, y_train)

    val_preds = clf.predict(X_val)
    acc = accuracy_score(y_val, val_preds)
    f1 = f1_score(y_val, val_preds)
    prec = precision_score(y_val, val_preds)
    rec = recall_score(y_val, val_preds)

    err = per_stratum_error(clf, val_df)
    err_s = pd.Series([err[s] for s in unique_strata], index=unique_strata)

    alpha = 0.85
    beta = 0.15
    inv_freq = 1.0 / strata_counts
    inv_freq_norm = inv_freq / inv_freq.sum()
    err_norm = err_s / err_s.sum()
    new_weights = alpha * err_norm + beta * inv_freq_norm
    momentum = 0.6
    updated = (1 - momentum) * weights + momentum * new_weights
    updated = updated / updated.sum()
    weights = updated.copy()

    history["epoch"].append(epoch)
    history["overall_acc"].append(acc)
    history["overall_f1"].append(f1)
    history["per_stratum_error"].append(err.copy())
    history["weights"].append(weights.copy())

    print(f"Epoch {epoch:02d}: acc={acc:.4f}, f1={f1:.4f}, precision={prec:.4f}, recall={rec:.4f}")
    print("  per-stratum error:", {k: round(v, 3) for k, v in err.items()})
    print("  updated weights:", {int(k): round(float(weights[k]), 3) for k in weights.index})
    print("-" * 72)

# Visualization
hist_df = pd.DataFrame({"epoch": history["epoch"], "accuracy": history["overall_acc"], "f1": history["overall_f1"]})

err_matrix = pd.DataFrame(history["per_stratum_error"]).T
err_matrix.columns = [f"epoch_{e}" for e in history["epoch"]]
err_matrix.index = [f"stratum_{s}" for s in err_matrix.index]

weights_matrix = pd.DataFrame(history["weights"]).T
weights_matrix.columns = [f"epoch_{e}" for e in history["epoch"]]
weights_matrix.index = [f"stratum_{s}" for s in weights_matrix.index]

plt.figure(figsize=(8, 4))
plt.plot(hist_df["epoch"], hist_df["accuracy"], marker='o')
plt.plot(hist_df["epoch"], hist_df["f1"], marker='s')
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Overall validation metrics across ASS epochs")
plt.legend(["Accuracy", "F1"])
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8, 4))
for s in err_matrix.index:
    plt.plot(range(1, err_matrix.shape[1]+1), err_matrix.loc[s].values, marker='o', label=s)
plt.xlabel("Epoch")
plt.ylabel("Validation Error (per stratum)")
plt.title("Per-stratum validation error evolution")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8, 4))
for s in weights_matrix.index:
    plt.plot(range(1, weights_matrix.shape[1]+1), weights_matrix.loc[s].values, marker='o', label=s)
plt.xlabel("Epoch")
plt.ylabel("Sampling Weight (normalized)")
plt.title("Adaptive strata weights across epochs")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Print final tables
final_weights = pd.DataFrame({
    "stratum": unique_strata,
    "final_weight": [weights[s] for s in unique_strata],
    "train_count": [strata_counts[s] for s in unique_strata],
    "label_noise": [label_noise_by_stratum[s] for s in unique_strata]
}).set_index("stratum")

print("\nFinal strata weights and info:")
display(final_weights)

metrics_df = pd.DataFrame({
    "epoch": history["epoch"],
    "accuracy": history["overall_acc"],
    "f1": history["overall_f1"]
}).set_index("epoch")
print("\nPer-epoch overall metrics:")
display(metrics_df)

print("\nPer-stratum validation error (rows=strata, cols=epochs):")
display(err_matrix)
# Optional: save CSVs to download
final_weights.to_csv("final_strata_weights.csv")
metrics_df.to_csv("metrics_per_epoch.csv")
err_matrix.to_csv("per_stratum_error.csv")
print("\nCSV files saved: final_strata_weights.csv, metrics_per_epoch.csv, per_stratum_error.csv")
```

```python
# Demo: Adaptive Stratified Sampling (ASS)  toy implementation
# This notebook demonstrates the philosophy and mechanics of ASS by:
# 1. creating a synthetic tabular dataset with several strata (groups)
# 2. injecting different difficulty levels per stratum (via label noise)
# 3. implementing an adaptive stratified sampling loop:
#    - sample training data according to current strata weights
#    - train a classifier on that sampled data
#    - evaluate per-stratum error on a validation set
#    - update strata weights to emphasize strata with higher error
# 4. show how weights, per-stratum errors, and overall metrics evolve
#
# Notes:
# - Uses only scikit-learn, numpy, pandas, matplotlib (available in Colab)
# - Uses matplotlib (no seaborn) as requested
# - Uses ace_tools.display_dataframe_to_user to show tables where helpful
# - This is a concise educational demo, not a production-ready ASS implementation.

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
import matplotlib.pyplot as plt
import math
from collections import defaultdict
import random
import ace_tools

# 1) Create synthetic dataset
np.random.seed(42)
n = 8000

# features: 6 numeric features
X = np.random.normal(size=(n, 6))

# create 4 strata (groups) that represent e.g. demographic subgroups
# strata assignment is imbalanced intentionally
strata_probs = [0.6, 0.2, 0.15, 0.05]  # stratum 0 is majority, 3 is minority
strata = np.random.choice(len(strata_probs), size=n, p=strata_probs)

# create a true underlying risk score that depends on features and strata
# and add strata-specific difficulty via label noise
weights = np.array([0.5, -0.3, 0.2, 0.1, -0.2, 0.3])
base_score = X.dot(weights)

# strata-specific intercepts
strata_intercepts = np.array([-0.2, 0.1, 0.5, -0.5])
score = base_score + strata_intercepts[strata]

# convert to probabilities, then binary labels with threshold
probs = 1 / (1 + np.exp(-score))
y_clean = (probs > 0.5).astype(int)

# inject label noise differently per stratum to simulate "difficulty"
# higher noise => harder to learn => ASS should upweight that stratum
label_noise_by_stratum = {0: 0.02, 1: 0.05, 2: 0.12, 3: 0.25}  # increasing difficulty
y = y_clean.copy()
for s, noise in label_noise_by_stratum.items():
    idx = np.where(strata == s)[0]
    flip = np.random.rand(len(idx)) < noise
    y[idx[flip]] = 1 - y[idx[flip]]

# assemble DataFrame for convenience
df = pd.DataFrame(X, columns=[f"f{i}" for i in range(X.shape[1])])
df["stratum"] = strata
df["y"] = y

# train/val split that preserves strata distribution
train_df, val_df = train_test_split(df, stratify=df["stratum"], test_size=0.25, random_state=42)

print("Dataset composition (train):")
print(train_df["stratum"].value_counts(normalize=False).sort_index())

# 2) ASS algorithm (toy iterative reweighting)
# initialize strata weights (start with inverse frequency or uniform)
unique_strata = sorted(df["stratum"].unique())
n_strata = len(unique_strata)
strata_counts = train_df["stratum"].value_counts().sort_index()

# initialize weights: uniform
weights = pd.Series(1.0 / n_strata, index=unique_strata).astype(float)

# bookkeeping
history = {
    "epoch": [],
    "overall_acc": [],
    "overall_f1": [],
    "per_stratum_error": [],  # list of dicts
    "weights": []
}

# helper to sample according to strata-level weights (weighted stratified sampling)
def sample_train_by_weights(train_df, strata_weights, sample_size=None, random_state=None):
    """
    strata_weights: Series indexed by stratum -> relative weight (not necessarily summing to 1)
    This function returns a sampled DataFrame with replacement across strata according to desired
    proportion of total sample size. We compute desired per-stratum counts proportional to
    strata_weights * original_stratum_count to avoid overweighting extremely small strata.
    """
    if sample_size is None:
        sample_size = len(train_df)
    rng = np.random.RandomState(random_state)
    # compute per-stratum base counts = original counts
    base_counts = train_df["stratum"].value_counts().sort_index()
    # desired share ~ strata_weights * base_counts (so both weight and availability matter)
    desired_score = strata_weights * base_counts
    desired_share = desired_score / desired_score.sum()
    desired_counts = np.round(desired_share * sample_size).astype(int)
    # ensure at least 1 sample per stratum to avoid empty picks
    for s in unique_strata:
        if desired_counts[s] < 1:
            desired_counts[s] = 1
    # sample with replacement within each stratum
    pieces = []
    for s in unique_strata:
        subset = train_df[train_df["stratum"] == s]
        k = desired_counts[s]
        chosen = subset.sample(n=k, replace=True, random_state=rng.randint(0, 1_000_000))
        pieces.append(chosen)
    sampled = pd.concat(pieces).sample(frac=1.0, random_state=rng.randint(0, 1_000_000)).reset_index(drop=True)
    return sampled

# function to compute per-stratum error on validation set
def per_stratum_error(clf, val_df):
    preds = clf.predict(val_df[[c for c in val_df.columns if c.startswith("f")]])
    dfp = val_df.copy()
    dfp["pred"] = preds
    err = dfp.groupby("stratum").apply(lambda g: (g["pred"] != g["y"]).mean()).to_dict()
    return err

# run iterative adaptation for a number of epochs
EPOCHS = 12
for epoch in range(1, EPOCHS + 1):
    # 1) sample training data according to current strata weights
    sampled_train = sample_train_by_weights(train_df, weights, sample_size=len(train_df), random_state=epoch*13)
    X_train = sampled_train[[c for c in sampled_train.columns if c.startswith("f")]].values
    y_train = sampled_train["y"].values
    X_val = val_df[[c for c in val_df.columns if c.startswith("f")]].values
    y_val = val_df["y"].values

    # 2) train a classifier (we retrain from scratch each epoch in this toy demo)
    clf = RandomForestClassifier(n_estimators=80, random_state=epoch*7, n_jobs=-1)
    clf.fit(X_train, y_train)

    # 3) evaluate overall metrics on validation set
    val_preds = clf.predict(X_val)
    acc = accuracy_score(y_val, val_preds)
    f1 = f1_score(y_val, val_preds)
    prec = precision_score(y_val, val_preds)
    rec = recall_score(y_val, val_preds)

    # 4) compute per-stratum error and update strata weights
    err = per_stratum_error(clf, val_df)
    # convert to Series aligned with unique_strata order
    err_s = pd.Series([err[s] for s in unique_strata], index=unique_strata)

    # update rule: new_weight = alpha * normalized_error + beta * (1 / base_count_norm)
    # This blends error-driven emphasis with inverse-frequency to not fully ignore small strata.
    alpha = 0.85
    beta = 0.15
    inv_freq = 1.0 / strata_counts
    inv_freq_norm = inv_freq / inv_freq.sum()
    err_norm = err_s / err_s.sum()
    new_weights = alpha * err_norm + beta * inv_freq_norm
    # smooth update (momentum-like) to avoid large jumps
    momentum = 0.6
    updated = (1 - momentum) * weights + momentum * new_weights
    # normalize to sum to 1 for interpretability
    updated = updated / updated.sum()
    weights = updated.copy()

    # record
    history["epoch"].append(epoch)
    history["overall_acc"].append(acc)
    history["overall_f1"].append(f1)
    history["per_stratum_error"].append(err.copy())
    history["weights"].append(weights.copy())

    print(f"Epoch {epoch:02d}: acc={acc:.4f}, f1={f1:.4f}, precision={prec:.4f}, recall={rec:.4f}")
    print("  per-stratum error:", {k: round(v, 3) for k, v in err.items()})
    print("  updated weights:", {int(k): round(float(weights[k]), 3) for k in weights.index})
    print("-" * 72)

# 3) Visualize results

# convert history to dataframe for plotting
hist_df = pd.DataFrame({
    "epoch": history["epoch"],
    "accuracy": history["overall_acc"],
    "f1": history["overall_f1"]
})

# prepare per-stratum error and weights matrix
err_matrix = pd.DataFrame(history["per_stratum_error"]).T
err_matrix.columns = [f"epoch_{e}" for e in history["epoch"]]
err_matrix.index = [f"stratum_{s}" for s in err_matrix.index]

weights_matrix = pd.DataFrame(history["weights"]).T
weights_matrix.columns = [f"epoch_{e}" for e in history["epoch"]]
weights_matrix.index = [f"stratum_{s}" for s in weights_matrix.index]

# plot overall metrics
plt.figure(figsize=(8, 4))
plt.plot(hist_df["epoch"], hist_df["accuracy"], marker='o')
plt.plot(hist_df["epoch"], hist_df["f1"], marker='s')
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Overall validation metrics across ASS epochs")
plt.legend(["Accuracy", "F1"])
plt.grid(alpha=0.3)
plt.show()

# plot per-stratum error evolution
plt.figure(figsize=(8, 4))
for s in err_matrix.index:
    plt.plot(range(1, err_matrix.shape[1]+1), err_matrix.loc[s].values, marker='o', label=s)
plt.xlabel("Epoch")
plt.ylabel("Validation Error (per stratum)")
plt.title("Per-stratum validation error evolution")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# plot strata weight evolution
plt.figure(figsize=(8, 4))
for s in weights_matrix.index:
    plt.plot(range(1, weights_matrix.shape[1]+1), weights_matrix.loc[s].values, marker='o', label=s)
plt.xlabel("Epoch")
plt.ylabel("Sampling Weight (normalized)")
plt.title("Adaptive strata weights across epochs")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# display final tables using ace_tools display helper
final_weights = pd.DataFrame({
    "stratum": unique_strata,
    "final_weight": [weights[s] for s in unique_strata],
    "train_count": [strata_counts[s] for s in unique_strata],
    "label_noise": [label_noise_by_stratum[s] for s in unique_strata]
}).set_index("stratum")

ace_tools.display_dataframe_to_user("Final Strata Weights and Info", final_weights)

# show per-epoch overall metrics
metrics_df = pd.DataFrame({
    "epoch": history["epoch"],
    "accuracy": history["overall_acc"],
    "f1": history["overall_f1"]
}).set_index("epoch")
ace_tools.display_dataframe_to_user("Per-epoch Overall Metrics", metrics_df)

# show per-stratum errors as a table
err_table = err_matrix.copy()
err_table.index.name = "stratum"
ace_tools.display_dataframe_to_user("Per-stratum validation error (rows=strata, cols=epochs)", err_table)

print("Demo complete. You can inspect the plots and tables above to see how ASS shifts sampling towards strata with higher validation error (harder strata).")

```
